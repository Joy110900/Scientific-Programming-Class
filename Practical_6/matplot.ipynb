# Matplotlib

```python
#Importing required libraries
import matplotlib.pyplot as plt
import random as r
```

```python
#Generating random runs scored by player in each team
India = [r.randint(0, 100) for i in range(10)]
England = [r.randint(0, 100) for i in range(10)]
#Initialising a list with wickets from 1-10 and adding 11 to show the end
Wickets = [i for i in range(1, 12)]
```

```python
#Random runs generated
print("India  -", India)
print("England -", England)
print("Wickets -",Wickets)
```

```python
#Counting total runs after each wicket to plot
Iruns = []
Eruns = []
ti = 0
te = 0
for i in range(len(India)):
    Iruns.append(India[i] + ti)
    ti += India[i]
    Eruns.append(England[i] + te)
    te += England[i]

#Appending 0 to mark an end to inning
Iruns.append(0)
Eruns.append(0)

print(Iruns)
print(Eruns)
```

```python
#Plotting total runs scored by each team and plotting them against each other for comparative analysis
plt.plot(Wickets, Iruns, color='blue', marker='o', label='India') #Plot for runs scored by India vs fall of wicket
plt.plot(Wickets, Eruns, color='red', marker='o', label='England') #Plot for runs scored by England vs fall of wicket
plt.title('Performance') #Adding title to our plot
#Adding x and y labels
plt.xlabel('Wickets')
plt.ylabel('Runs')
#Adding legend to the plot 
plt.legend()
plt.show()
```
