```python
import matplotlib.pyplot as plt
import random as r
```

```python
India = [r.randint(0, 100) for i in range(10)]
England = [r.randint(0, 100) for i in range(11)]
Wickets = [i for i in range(1, 12)]
```

```python
print("India  -", India)
print("England -", England)
print("Wickets -",Wickets)
```

```python
Iruns = []
Eruns = []
ti = 0
te = 0
for i in range(len(India)):
    Iruns.append(India[i] + ti)
    ti += India[i]
    Eruns.append(England[i] + te)
    te += England[i]

Iruns.append(0)
Eruns.append(0)

print(Iruns)
print(Eruns)
```

```python
plt.plot(Wickets, Iruns, color='blue', marker='o', label='India')
plt.plot(Wickets, Eruns, color='red', marker='o', label='England')
plt.title('Performance')
plt.xlabel('Wickets')
plt.ylabel('Runs')
plt.legend()
plt.show()
```
